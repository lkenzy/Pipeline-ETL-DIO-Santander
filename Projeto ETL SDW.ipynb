    Explicação: Através de um Dataset do Kaggle, criei um ETL que como objetivo final 
    devolve gráficos mostrando as médias salariais nos Estados Unidos de 1973 a 2022 de acordo com o sexo e cor.
    
    LINK DO DATASET: https://www.kaggle.com/datasets/asaniczka/median-and-avg-hourly-wages-in-the-usa-1973-2022/

#### Extract:

```python
import pandas as pd

df = pd.read_csv("base_dados.csv")
df
```

#### Transform:

```python
for column in df.columns:
    if "median" in column:
        df = df.drop(columns=[column])
colunmn_to_remove_from = "recent_high-school_graduate" # a partir de qual coluna quero remover
start_column_index = df.columns.get_loc(colunmn_to_remove_from) # obtendo o indice desta coluna
df = df.iloc[:, :start_column_index] # removendo todas as colunas sucessoras
df
```

```python
for column in df.columns:
    if "hispanic" in column:
        df = df.drop(columns=[column])
df
```

```python
df = df.drop("average", axis= 1)
df
```

```python
columns_to_process = df.columns[df.columns != 'year'] # uma lista com todas as colunas, menos a "year"

dataframes = {} 

for column in columns_to_process:
    dataframes[column] = df[['year', column]]
dataframes["men_average"] # agora cada coluna tem seu df, basta mudar o nome da chave
```

#### Load:

```python
import plotly.express as px
import plotly.subplots
import plotly.graph_objs as go

colors = ["red", "yellow", "green"]
column_names = {
    'men_average': 'Para Homem',
    'women_average': 'Para Mulher',
    'white_average': "Para Branco",
    'black_average': "Para Preto",
    'white_men_average': "Para Homem Branco",
    'black_men_average': "Para Homem Preto",
    'white_women_average': "Para Mulher Branca",
    'black_women_average': "Para Mulher Preta"
}

# lista de cores correspondente às colunas
line_colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan']

# Crie subplots com duas colunas por linha
num_rows = (len(dataframes) + 1) // 2
num_cols = 2
fig = plotly.subplots.make_subplots(rows=num_rows, cols=num_cols, subplot_titles=list(column_names.values()))

row, col = 1, 1

for column, dataframe in dataframes.items():
    column_name = column_names.get(column, column)
    line_color = line_colors[list(column_names.keys()).index(column)]
    fig.add_trace(go.Scatter(x=dataframe['year'], y=dataframe[column], mode='lines', line=dict(color=line_color), name=column_name), row=row, col=col)
    
    if col < 2:
        col += 1
    else:
        col = 1
        row += 1

# Atualize o layout dos subplots
fig.update_layout(width=800, height=500 * num_rows, paper_bgcolor='white', plot_bgcolor='white', showlegend=False)

# bExiba os subplots
fig.show()


```
